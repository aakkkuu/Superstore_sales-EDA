Descriptive statistic for key metrics

```python
# Convert the 'Order Date' to datetime format
data_df['Order Date'] = pd.to_datetime(data_df['Order Date'], format="%d/%m/%Y")

# Calculate the 'Total_Sales'
total_sales = data_df['Sales'].sum()

# Calculate the 'Average Order Sales'
average_order_sales = data_df.groupby('Order ID')['Sales'].sum().mean()

# Print the results
print("Total Sales:", total_sales)
print("Average Order Value:", average_order_sales)
```

```python
# Plot the distribution of the sales
plt.figure(figsize = (10,6))
sns.histplot(data_df['Sales'], kde = True, bins =30, color='blue')
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plot the distribution of the category
plt.figure(figsize = (10,6))
sns.histplot(data_df['Category'], kde = True, bins =30, color='blue')
plt.title('Distribution of Category')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plot the distribution of the segment
plt.figure(figsize = (10,6))
sns.histplot(data_df['Segment'], kde = True, bins =30, color='blue')
plt.title('Distribution of Segment')
plt.xlabel('Segment')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Customer Segmentation

```python
# Types of Customer
types_of_customer = data_df['Segment'].unique()
print(types_of_customer)
```

```python
# Numbers of customer in each segment
customer_segment = data_df['Segment'].value_counts().reset_index()
customer_segment = customer_segment.rename(columns = {'Segment':'Customer Type', 'count':'Total Customers'})
print(customer_segment)
```

```python
print(customer_segment.columns)
```

```python
# Define custom colors for the pie chart
# Explode the first slice (Type A)
explode = (0.1, 0, 0)

# Color palette
colors = ['#ff9999','#66b3ff','#99ff99']

# Plot
plt.figure(figsize=(8, 6))
plt.pie(customer_segment['Total Customers'], labels=customer_segment['Customer Type'], autopct='%1.1f%%', startangle=90, colors=colors, explode=explode)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

plt.title('Distribution of Customer Segments')
plt.show()
```

```python
# Group the data by segment and  then calculate the total sales for each segment
sales_by_segment= data_df.groupby('Segment')['Sales'].sum().reset_index()
sales_by_segment = sales_by_segment.rename(columns = {'Segment':'Customer Type','Sales':'Total Sales'})
print(sales_by_segment)
```

```python
#Ploting a bar graph
plt.figure(figsize=(8, 6))
plt.bar(sales_by_segment['Customer Type'], sales_by_segment['Total Sales'], color='skyblue')

# Add title and labels
plt.title('Sales By Segment')
plt.xlabel('Customer Type')
plt.ylabel('Total Sales')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

High value Customer

```python
# Group by customer ID, name, and segment, then sum up the sales
customer_spending = data_df.groupby(['Customer ID', 'Customer Name', 'Segment'])['Sales'].sum().reset_index()

# Sort the customers based on total sales in descending order
top_spend = customer_spending.nlargest(5, 'Sales').reset_index(drop=True)

# Print the result
print(top_spend)
```

Purchasing Behaviour of Customer

```python
# Group the data by Customer ID, Customer Name, and Segment, and calculate the frequency for each customer
customer_order_frequency = data_df.groupby(['Customer ID', 'Customer Name', 'Segment'])['Order ID'].count().reset_index()

# Rename the column for Order ID to Total Orders
customer_order_frequency.rename(columns={'Order ID': 'Total Orders'}, inplace=True)

# Identify repeat customers and sort them based on descending order frequency
repeat_customers_sorted = customer_order_frequency[customer_order_frequency['Total Orders'] >= 2]\
                                                .sort_values(by='Total Orders', ascending=False)\
                                                .head(10)\
                                                .reset_index(drop=True)

# Print the repeat customers with high order frequency
print(repeat_customers_sorted)
```

```python
# Analyze characteristics for each customer segment
customer_segments = data_df['Segment'].unique()

for segment in customer_segments:
    segment_data = data_df[data_df['Segment'] == segment]

    # Basic summary statistics
    print("Summary statistics for", segment, "segment:")
    print(segment_data.describe())

    # Visualize total sales by category
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Category', y='Sales', data=segment_data)
    plt.title('Total Sales by Category for ' + segment + ' segment')
    plt.xlabel('Category')
    plt.ylabel('Total Sales')
    plt.xticks(rotation=45)
    plt.show()

    # Visualize distribution of sales over time
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='Order Date', y='Sales', data=segment_data)
    plt.title('Distribution of Sales Over Time for ' + segment + ' segment')
    plt.xlabel('Order Date')
    plt.ylabel('Sales')
    plt.xticks(rotation=45)
    plt.show()


```

Top selling product analysis

```python
# Types of sub-category
sub_category = data_df['Sub-Category'].unique
print(sub_category)
```

```python
# Grouping the data based on the cateogry with sub-category in each of them
sub_categories = data_df.groupby('Category')['Sub-Category'].nunique().reset_index()

# Sort the sub_categories
sub_categories = sub_categories.sort_values(by='Sub-Category', ascending = False)

# Print the sub category
print(sub_categories)
```

```python
# Sales of each sub-category
sale_by_category = data_df.groupby(['Category','Sub-Category'])['Sales'].sum().reset_index()

# Sort the DataFrame by the 'Sales' column in ascending order
sale_by_category = sale_by_category.sort_values(by ='Sales', ascending=False)

top_selling_product = sale_by_category.reset_index(drop=True)

print(top_selling_product)
```

```python
plt.figure(figsize=(8, 6))
plt.barh(top_selling_product['Sub-Category'], top_selling_product['Sales'], color='skyblue')
plt.xlabel('Sales')
plt.ylabel('Sub-Category')
plt.title('Top Selling Products by Sub-Category')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest sales at the top
plt.show()
```

```python
# Grouping the product category by sales
product_category = data_df.groupby('Category')['Sales'].sum().reset_index()

# Sort the category based on sales
top_product_category = product_category.sort_values(by='Sales',ascending = False)

print(top_product_category)
```

```python
# Data preparation
labels = top_product_category['Category']
sales = top_product_category['Sales']

# Plotting the pie chart
plt.figure(figsize=(6, 4))
plt.pie(sales, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Product Category')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Product Performance over Time

```python
# Data Preparation
data_df['Order Date'] = pd.to_datetime(data_df['Order Date'])
data_df['Year'] = data_df['Order Date'].dt.year

# Grouping and Aggregating
sub_category_performance = data_df.groupby(['Year', 'Category'])['Sales'].sum().reset_index()

# Visualization
plt.figure(figsize=(10, 8))
for sub_category, sub_df in sub_category_performance.groupby('Category'):
    plt.plot(sub_df['Year'], sub_df['Sales'], label=sub_category)

plt.title('Product Performance Over Time')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

Sales trend over different time period

```python
# Convert the 'Order_date' to datetime format
data_df['Order Date'] = pd.to_datetime(data_df['Order Date'])

# Group the data by months and calculate the total sales over month period
monthly_sale = data_df.resample('M', on='Order Date')['Sales'].sum()

# Plot the sales by month
plt.figure(figsize=(10, 6))
monthly_sale.plot(marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.grid(True)  # Add grid lines for better readability
plt.show()
```

```python
# Convert the 'Order_date' to datetime format
data_df['Order Date'] = pd.to_datetime(data_df['Order Date'])

# Group the data by quaterly and calculate the total sales over month period
monthly_sale = data_df.groupby(data_df['Order Date'].dt.to_period('Q'))['Sales'].sum()

# Plot the sales by month
plt.figure(figsize = (10,6))
plt.subplot(1, 1, 1)
monthly_sale.plot(kind='line', marker='o')
plt.title('Quaterly Sales Trend')
plt.xlabel('Quaterly')
plt.ylabel('Sales Amount')
plt.show()
```

```python
# Convert the 'Order_date' to datetime format
data_df['Order Date'] = pd.to_datetime(data_df['Order Date'])

# Group the data by yearly and calculate the total sales over Year period
monthly_sale = data_df.groupby(data_df['Order Date'].dt.to_period('Y'))['Sales'].sum()

# Plot the sales by month
plt.figure(figsize = (10,6))
plt.subplot(1, 1, 1)
monthly_sale.plot(kind='line', marker='o')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales Amount')
plt.show()
```
